```python
# Determine if a year is a leap year using nested condition statement

year= int(input("enter a year"))

if year%4==0:
  if year%100==0:
    if year%400==0:
      print(year,"is a leap year")
    else:
      print(year,'is not a leap year')
  else:
      print(year,'is a leap year')
else:
  print(year,'is not a leap year')
```

```python
## Assignment
## Simple calculator program
num1 = int(input("enter a number:"))
num2 = int(input("enter a number "))
operation = input("Enter a operator (+,-,*,/):")

if operation == '+':
  result = num1+num2
elif operation == '-':
  result = num2 - num1
elif operation == '*':
  result = num1 * num2
elif operation == '/':
  if num2 != 0:
    result = num1 / num2
  else:
    result = "Error! Divison by Zero."

else:
  result = "Invalid operation."

print("Result:", result )

```

```python
# EX-1  Manage a TO do list

to_do_lst=['buy Gold','cleaning','pay bills']

## adding to task
to_do_lst.append("Schedule meeting")
to_do_lst.append("go for a run")

## Removing a completed task
to_do_lst.remove('cleaning')

## check if a task is in the list

if 'pay bills'  in to_do_lst:
  print("dont forget to pay the bills")


print("TO do lst remaining :-")
for task in to_do_lst:
  print(f"{task}")
```

```python
## EX 2 Organing the student grade
grades=[85,92,78,90,88]

## adding new grade
grades.append(95)

##Calculating the avarage grade
average_grade= sum(grades)/ len(grades)
print(f"Average Grade:{average_grade:.2f}")

## finding the highest and lowest grades
highest_grade = max(grades)
lowest_grade = min(grades)
print(f"Highest_grade: {highest_grade}")
print(f"lowest_grade: {lowest_grade}")
```

```python
#EX-3 Managing an inventory
inventory = ["apples","bananas","oranges","grapes"]

#adding a new item
inventory.append("strawberries")

# Removing a item that is out of stock
inventory.remove("bananas")

# checking if an item is in stock
item = "oranges"
if item in inventory:
  print(f"{item} are in stock.")
else:
  print(f"{item} are out of stock.")

  ## printing the inventory
print("Inventory list:")
for item in inventory:
 print(f" - {item}")
```

```python
## EX-4 Collecting user feedback
feedback =['great service!','very satisfied','could be  better','excellent experience']

# adding new feedback
feedback.append("Not happy with the service")

# Counting specific feedback
positive_feedback_count = sum(1 for comment in feedback if 'great' in comment.lower() or 'excellent'in comment.lower())
print(f"Positive feedback count:{positive_feedback_count}")

# Printing all feedback
print("User feedback:")
for comment in feedback:
  print(f"-{comment}")
```
